```python
import numpy as np
```

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])
gates={'AND':[0,0,0,1],
   'OR':[0,1,1,1],
   'NAND':[1,1,1,0],
   'NOR':[1,0,0,0],
   'XOR':[0,1,1,0]}
```

```python
class perceptron:
  def __init__(self,lr,iter):
    self.lr=lr
    self.epoch=iter
    self.weight=None
    self.bias=None

  def fit(self,x,y):
     self.weight=np.zeros(x.shape[1])
     self.bias=0

     for e in range(self.epoch):
       for i in range(x.shape[0]):
         y_pred=self.afunc(np.dot(self.weight,x[i])+self.bias)
         mae=y[i]-y_pred
         self.weight=self.weight+self.lr*mae*x[i]
         self.bias=self.bias+self.lr*mae

  def afunc(self,activation):
    if(activation>=0):
      return 1
    else:
      return 0

  def predict(self,x):
    preds=[]
    for i in range(x.shape[0]):
        l=np.dot(self.weight,x[i])+self.bias
        preds.append(self.afunc(l))
    return np.array(preds)
```

```python
#AND Gate

And=perceptron(0.1,10)
And.fit(x,gates['AND'])
```

```python
print('Final Weights:',And.weight)
print('Final Bias:',And.bias)
print('Predictions:',And.predict(x))

if np.array_equal(And.predict(x),gates['AND']):
  print('Learned Successfully')
else:
  print('Misclassifications')
```
