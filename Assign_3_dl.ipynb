```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("multiple_linear_regression_dataset.csv")
```

```python
data.head()
```

```python
data.shape,data.columns
```

```python
y=data['income'].values
x=data.drop('income',axis=1).values
```

```python
x.shape,y.shape
```

```python
w=np.zeros(x.shape[1])
b=0.0
```

```python
def predict(x,w,b):
  y_pred=np.dot(x,w)+b
  return y_pred
```

```python
def mse(y,y_pred):
  loss=((y_pred-y)**2).mean()
  return loss
```

```python
def update(x,y,y_pred,w,b,lr):
  dw=(2/len(y))*np.dot(x.T,(y_pred-y))
  db=(2/len(y))*np.sum(y_pred-y)
  w=w-lr*dw
  b=b-lr*db
  return w,b
```

```python
lr=0.0001
epoch=1000
for e in range(epoch):
  y_pred=predict(x,w,b)
  loss=mse(y,y_pred)
  w,b=update(x,y,y_pred,w,b,lr)

  if e%100==0:
    print('Epoch :',e,',Weights :',w,',Bias :',b,'Loss :',loss,'\n')
```

```python
print('Final Weights :',w)
print('Final Bias :',b)
```

```python
a=np.array([68,4.5])
salary=predict(a,w,b)
print('Salary of candidate with age=68 and years of experience=4.5 is',salary)
```
