```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('abalone.data',header=None)
```

```python
data.columns=['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']
```

```python
data.head()
```

```python
data.shape[0]     #No. of rows
```

```python
data.columns
```

```python
data['y']=data['Rings']+1.5
data.head()
```

```python
x=np.column_stack((data['Length']*data['Height'],data['Diameter'],data['Shell weight']))
y=data['y'].values
```

```python
#Feature 1: (Length*Height)
#This feature approximates body volume contribution.
#Since abalone age is related to growth, a size-based interaction

#Feature 2: (Diameter)
#Represents overall body width.
#Together with length,it reflects general body size and growth stage.

#Feature 3: (Shell weight)
#Shell weight increases as the abalone ages.
```

```python
s=int(0.8*len(x))
x_train=x[:s]
y_train=y[:s]
x_test=x[s:]
y_test=y[s:]

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
mean=x_train.mean(axis=0)
std=x_train.std(axis=0)
print('Mean =',mean,',std =',std)
```

```python
#Normalisation
x_train=(x_train-mean)/std
x_test=(x_test-mean)/std
```

```python
w=np.zeros(x_train.shape[1])
b=0.0

def forward(x,w,b):
    y_pred=np.dot(x,w)+b
    return y_pred

#parameters are w,b
#No. of parameters 4
```

```python
y_pred=forward(x_train, w, b)
x_train.shape,w.shape,y_pred.shape
```

```python
def mse(y,y_pred):
  loss=np.mean((y-y_pred)**2)
  return loss
```

```python
def grad_dw(x,y,y_pred):
  dw=(2/len(y))*np.dot(x.T,(y_pred-y))
  return dw

def grad_db(y,y_pred):
  db=(2/len(y))*np.sum(y_pred-y)
  return db
```

```python
lr=0.001
epoch=1000

for e in range(epoch):
  y_pred=forward(x_train,w,b)
  loss=mse(y_train,y_pred)
  dw=grad_dw(x_train,y_train,y_pred)
  db=grad_db(y_train,y_pred)
  w=w-lr*dw
  b=b-lr*db

  if e%100==0:
    print("Epoch :",e,"Loss :",loss)
```

```python
y_test_pred=forward(x_test,w,b)
print('MSE for test data : ',mse(y_test,y_test_pred))
print('MAE for test data : ',np.mean(np.abs(y_test-y_test_pred)))
```

```python
for i in range(5):
  print('True Age :',y_test[i])
  print('Predicted Age :',y_test_pred[i])
  print('Error :',y_test[i]-y_test_pred[i],'\n')
```
