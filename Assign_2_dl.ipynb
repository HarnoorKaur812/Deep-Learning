```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("glass.csv")
```

```python
df.head()
```

```python
df.shape,df.columns
```

```python
df['Type'].unique()
```

```python
df['y']=(df['Type']==1).astype(int)
```

```python
df['y'].unique()
```

```python
df=df.drop(columns=['Type'])
```

```python
df.head()
```

```python
x=df.drop(columns=['y']).values
y=df['y'].values
x,y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=15)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
def sigmoid(a):
  return 1/(1+np.exp(-a))
```

```python
def predict_prob(x,w,b):
  a=np.dot(x,w)+b
  return sigmoid(a)
```

```python
def loss(y,p):
  return -np.mean(y*np.log(p)+(1-y)*np.log(1-p))
```

```python
def update(x,y,w,b,lr):
  p=predict_prob(x,w,b)
  e=p-y
  w=w-lr*np.dot(x.T,e)/len(y)
  b=b-lr*np.mean(e)
  return w,b
```

```python
w=np.zeros(x_train.shape[1])
b=0.0
lr=0.1
epoch=100
```

```python
for e in range(epoch):
  w,b=update(x_train,y_train,w,b,lr)
  if e%20==0:
    print('Epoch :',e,'\nWeight :',w,'\nBias :',b,'\n\n')
```

```python
def predict_label(p,threshold):
  return (p>=threshold).astype(int)
```

```python
p=predict_prob(x_test,w,b)
```

```python
#Threshold=0.5

y_pred=predict_label(p,0.5)
acc=np.mean(y_pred==y_test)
print('Accuracy with threshold=0.5 :',acc)
```

```python
#Threshold=0.7

y_pred=predict_label(p,0.7)
acc=np.mean(y_pred==y_test)
print('Accuracy with threshold=0.7 :',acc)
```
